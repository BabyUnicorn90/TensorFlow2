## 3.4.1 그래프를 그리는 함수 사용법 익히기

```python
#3.31 간단한 꺾은선그래프 그리기 
import tensorflow as tf
import matplotlib.pyplot as plt

x = range(20)
y = tf.random.normal([20], 0, 1)

plt.plot(x, y) 
#첫번째 인자: x축에 넣을 요소
#두번째 인자: y축에 넣을 요소
plt.show()
#plt.show() 함수를 호출해야만 주피터노트북이나 코랩에서 확인 가능. 
```

```python
#3.32 간단한 점그래프 그리기
plt.plot(x, y, 'bo')
plt.show()

plt.plot(x, y, 'b-')
plt.show()

plt.plot(x, y, 'b--')
plt.show()


plt.plot(x, y, 'ro')
plt.show()
```

```python
#3.33 정규분포 그래프를 히스토그램으로 나타내기 

random_normal = tf.random.normal = ([100000], 0, 1)
plt.hist(random_normal, bins=100)
plt.show()
```

## 3.4.2 2-레이어 XOR네트워크의 정보 시각화

학습이 잘 되고 있는지 확인하기 위한 측정치(metric)변화량을 나타내는 [선그래프]

(분류 등에서는 정확도(accuracy)등 다른 측정치도 관찰해야 함)



```python
#선 그래프를 이용해 손실이 어떻게 변했는지 알아보기
import numpy as np

x = np.array([[1, 1], [1, 0], [0, 1], [0, 0]])
y = np.array([[0], [1], [1], [0]])

model = tf.keras.Sequential([
    tf.keras.layers.Dense(units = 2, activation = 'sigmoid', input_shape = (2, )),
    tf.keras.layers.Dense(units = 1, activation = 'sigmoid')
])

model.compile(optimizer = tf.keras.optimizers.SGD(lr = 0.1), loss = 'mse')

history = model.fit(x, y, epochs=2000, batch_size = 1)

plt.plot(history.history['loss'])
#plt.plot()에 하나의 변수만 전달하면: 
# 그 변수를 y로 간주, x는 자동으로 reange(len(y))
```
